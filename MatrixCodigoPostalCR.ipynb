```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

df = pd.DataFrame()
headers = {
    'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10.12; rv:55.0) Gecko/20100101 Firefox/55.0',
}

URL = "https://codigo-postal.org/costa-rica/"
page = requests.get(URL, headers=headers)

soup = BeautifulSoup(page.content, "html.parser")
elementUL = soup.find("ul", {"class": "column-list"})
elementsLi = elementUL.find_all("li")
for li in elementsLi:
    href_provincia = li.find("a")["href"]
    page_provincia = requests.get(href_provincia, headers=headers)
    soup_provincia = BeautifulSoup(page_provincia.content, "html.parser")
    elementUL_cantones = soup_provincia.find("ul", {"class": "column-list"})
    elementsLi_cantones = elementUL_cantones.find_all("li")
    for li_canton in elementsLi_cantones:
        href_canton = li_canton.find("a")["href"]
        page_canton = requests.get(href_canton, headers=headers)
        soup_canton = BeautifulSoup(page_canton.content, "html.parser")
        
        table_distritos = soup_canton.find('table')
        table_rows = table_distritos.find_all('tr')
        
        l = []
        for tr in table_rows:
            td = tr.find_all('td')
            row = [tr.text for tr in td]
            l.append(row)
        df_distrito = pd.DataFrame(l, columns=["Provincia", "Canton", "Distrito", "Codigo Postal"])
        df = pd.concat([df, df_distrito])
        
df
```
