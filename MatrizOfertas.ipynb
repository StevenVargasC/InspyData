```python
dfOfertas = dd.read_csv('D:/DataChallenge2024/BD/TBRelacional/ofertas.csv')
```

```python
#Se tuvo que limpiar la columna direccion, cedula proveedor, nombre proveedor
dfProveedores = dd.read_csv('D:/DataChallenge2024/BD/TBRelacional/proveedores2.csv', dtype={'codigo_postal': 'string'})
```

```python
dfRepresentantes = dd.read_csv('D:/DataChallenge2024/BD/TBRelacional/representantes.csv')
```

```python
dfLineasAdjudicadas = dd.read_csv('D:/DataChallenge2024/BD/TBRelacional/lineasadjudicadas.csv')
```

```python
dfProcedimientos = dd.read_csv('D:/DataChallenge2024/BD/TBRelacional/procedimientos.csv')
```

```python
dfLineasProc = dd.read_csv('D:/DataChallenge2024/BD/TBRelacional/lineasproc.csv')
```

```python
dfLineasOfertadas = dd.read_csv('D:/DataChallenge2024/BD/TBRelacional/lineasofertadas.csv')
```

```python
dfProcedOfertas = dd.merge(dfProcedimientos, dfOfertas, how="inner", on=["nro_sicop"])
dfProcedOfertas = dfProcedOfertas[dfProcedOfertas['fecha_invitacion'] > "2018-01-01"]
```

```python
# Existen proveedores con un mismo nombre pero con diferente cedula, es correcto?
dfProveedoresAgrupado = dfProveedores.groupby(['nombre_proveedor']).size().reset_index()
dfProveedoresAgrupado = dfProveedoresAgrupado.rename(columns={0: 'count'})
dfProveedoresAgrupado[(dfProveedoresAgrupado['count'] > 1)].compute()
```

```python
# Existen fechas de Presentacion de Oferta con valor Null, esto es correcto?
dfProcedOfertas[dfProcedOfertas["fecha_presenta_oferta"].isnull()].compute()
```

```python
# Existen tipos de ofertas con valores null y llamado Consorcio inscrito.
dfProcedOfertas[~dfProcedOfertas["tipo_oferta"].isin(['Acuerdo Consorcial', 'Conjunta', 'Individual', 'Puja subasta'])].compute()
```

```python
dfProcedOfertasLinOfertadas = dd.merge(dfProcedOfertas, dfLineasOfertadas, how='inner', on=["nro_oferta"])
```

```python
# Revisar: Hay tipos de moneda con el valor XXX o Null
dfProcedOfertasLinOfertadas[~dfProcedOfertasLinOfertadas["tipo_moneda"].isin(['CRC', 'USD', 'EUR', 'GBP'])].compute()
```

```python
# Revisar: Hay tipos de moneda con el valor XXX o Null
dfProcedOfertasLinOfertadas[~dfProcedOfertasLinOfertadas["tipo_moneda"].isin(['CRC', 'USD', 'EUR', 'GBP']) & (dfProcedOfertasLinOfertadas["nro_procedimiento"] == '2024LE-000001-0018962008')].compute()
```

```python
# Existen Lineas Ofertadas que tienen como valor null el precio unitario ofertado aun cuando haya una cantidad ofertada
dfProcedOfertasLinOfertadas[dfProcedOfertasLinOfertadas["precio_unitario_ofertado"].isnull() & (dfProcedOfertasLinOfertadas["cantidad_ofertada"] != 0.0)].compute()
```

```python
dfProcedOfertasLinOfertadas[dfProcedOfertasLinOfertadas['nro_consorcio'].isnull()][['cedula_proveedor', 'nro_consorcio']].compute()
```

```python
dfProcedOfertasLinOfertadas[dfProcedOfertasLinOfertadas['nro_consorcio'].notnull()][['cedula_proveedor', 'nro_consorcio']].compute()
```
