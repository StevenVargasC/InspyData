```python
import pandas as pd
import numpy as np
from pathlib import Path 
```

```python
dfSolicitudes = pd.read_csv('DatosSitioWeb/tablas relacionadas/solicitudes.csv')
```

```python
dfProcedimientos = pd.read_csv('DatosSitioWeb/tablas relacionadas/procedimientos.csv')
```

```python
dfFuncionarioRol = pd.read_csv('DatosSitioWeb/tablas relacionadas/funcionarioRol.csv')
```

```python
dfLineasProc = pd.read_csv('DatosSitioWeb/tablas relacionadas/lineasproc.csv')
```

```python
dfInstituciones = pd.read_csv('DatosSitioWeb/tablas relacionadas/instituciones.csv')
```

```python
#Se tuvo que limpiar sector_insitucional
dfSectorInstituciones = pd.read_csv('DatosSitioWeb/tablas relacionadas/sector_institucional2.csv')
```

```python
dfAclaraciones = pd.read_csv('DatosSitioWeb/tablas relacionadas/aclaraciones.csv')
```

```python
dfInsitituciones_Sector = pd.merge(dfInstituciones, dfSectorInstituciones, how="inner", on=["cedula"])
```

```python
dfSolicitudesProcedimientos = pd.merge(dfSolicitudes, dfProcedimientos, how="inner", on=["nro_sicop"])
```

```python
dfSoliProceInsitu = pd.merge(dfSolicitudesProcedimientos, dfInsitituciones_Sector, left_on= ['cedula_institucion'],
right_on= ['cedula'], how="inner")
```

```python
dfLineasSoliProceInsitu = pd.merge(dfLineasProc, dfSoliProceInsitu, on=['nro_sicop'], how="inner")
```

```python
dfProceAclaraciones = pd.merge(dfLineasSoliProceInsitu, dfAclaraciones, on=['nro_sicop'], how="inner")
```

```python
dfProce = pd.merge(dfProceAclaraciones, dfFuncionarioRol, on=['nro_sicop'], how="inner")
```

```python
dfProce.head()
```

```python
dfProce.info()
```

```python
#Conversion automatica de datos
dfProce = dfProce.convert_dtypes()
```

```python
dfProce.info()
```

```python
#Convrsiones manuales booleanos
dfProce.cronograma = dfProce.cronograma == 'Y'
dfProce.costo_beneficio = dfProce.costo_beneficio == 'Y'
dfProce.pago_adelantado_pymes = dfProce.pago_adelantado_pymes == 'Y'

#Conversiones manuales Datetimes
dfProce.fecha_publicacion = pd.to_datetime(dfProce.fecha_publicacion,
               format='%Y-%m-%d %H:%M:%S')
dfProce.fecha_invitacion = pd.to_datetime(dfProce.fecha_invitacion,
               format='%Y-%m-%d %H:%M:%S')
dfProce.fecha_inicio_recepcion = pd.to_datetime(dfProce.fecha_inicio_recepcion,
               format='%Y-%m-%d %H:%M:%S')
dfProce.fecha_cierre_recepcion = pd.to_datetime(dfProce.fecha_cierre_recepcion,
               format='%Y-%m-%d %H:%M:%S')
dfProce.fecha_apertura = pd.to_datetime(dfProce.fecha_apertura,
               format='%Y-%m-%d %H:%M:%S')
dfProce.fecha_limite_recursos = pd.to_datetime(dfProce.fecha_limite_recursos,
               format='%Y-%m-%d %H:%M:%S')
dfProce.fecha_solicitud = pd.to_datetime(dfProce.fecha_solicitud,
               format='%Y-%m-%d %H:%M:%S')
dfProce.fecha_respuesta = pd.to_datetime(dfProce.fecha_respuesta,
               format='%Y-%m-%d %H:%M:%S')
               
               
         
              
```

```python
dfProce = dfProce[dfProce['fecha_invitacion'] > "2018-01-01"]
```
