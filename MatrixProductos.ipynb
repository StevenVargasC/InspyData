```python
import pandas as pd
import numpy as np
from pathlib import Path 
```

```python
dfJerarquiaClasificacionBS = pd.read_csv('DatosSitioWeb/tablas relacionadas/jerarquiaClasificacionBS.csv')
```

```python
dfClasificacionbs = pd.read_csv('DatosSitioWeb/tablas relacionadas/clasificacionbs.csv')
```

```python
dfIdentificacionbs = pd.read_csv('DatosSitioWeb/tablas relacionadas/identificacionbs.csv')
```

```python
dfProductobs = pd.read_csv('DatosSitioWeb/tablas relacionadas/productobs.csv')
```

```python
dfJerarquiaClasificacionBS = dfJerarquiaClasificacionBS.convert_dtypes()
```

```python
dfProductobs = dfProductobs.convert_dtypes()
```

```python
dfJerarquiaDetalleClasificacionBS = pd.merge(dfJerarquiaClasificacionBS, dfClasificacionbs, left_on= ['codigoClasificacion'],
right_on= ['codigo_clasificacion'], how="inner")
```

```python
dfJerarquiaDetalleClasificacionBS.head()  
```

```python
dfJerarquiaDetalleClasificacionBS.drop('codigoClasificacion', axis=1, inplace=True)
dfJerarquiaDetalleClasificacionBS.drop('nombreClasificacion', axis=1, inplace=True)
```

```python
dfJerarquiaDetalleClasificacionBS.head()
```

```python
dfClasificacionIdentificacion = pd.merge(dfJerarquiaDetalleClasificacionBS, dfIdentificacionbs, how="inner", on=["codigo_clasificacion"])
```

```python
dfClasificacionIdentificacion.head()
```

```python
dfProductosBS = pd.merge(dfClasificacionIdentificacion, dfProductobs, how="inner", on=["codigo_identificacion"])
```

```python
dfProductosBS.head()
```

```python
dfProductosBS.info()
```

```python
dfProductosBS.drop('codigo_clasificacion_p', axis=1, inplace=True)
dfProductosBS.drop('codigo_identificacion_p_y', axis=1, inplace=True)  
dfProductosBS.drop('aspectos_ambientales_sociales_x', axis=1, inplace=True)
```

```python
dfProductosBS.info()
```

```python
dfProductosBS.rename(columns={'codigo_identificacion': 'cod_clasi_identi', 'codigo_identificacion_p_x': 'codigo_identificacion', 'aspectos_ambientales_sociales_y': 'ambiente_social'}, inplace=True)
```

```python
dfProductosBS.head()
```

```python
dfProductosBS.info()
```

```python
dfProductosBS['intento'] = dfProductosBS['codigo_producto_p'].astype(str).apply(lambda x: x.zfill(8))
dfProductosBS
```

```python
dfProductosBS['compare'] = (dfProductosBS['codigo_producto'] == 
                            (dfProductosBS['cod_clasi_identi'].astype(str) + dfProductosBS['intento']))
```

```python
dfProductosBS
```

```python
value_counts = dfProductosBS['compare'].value_counts()
print(value_counts)
```

```python
dfProductosBS.drop('compare', axis=1, inplace=True)
dfProductosBS.drop('ambiente_social', axis=1, inplace=True)  
dfProductosBS.drop('intento', axis=1, inplace=True)
```

```python
filepath = Path('DatosSitioWeb/out/matrixProductos.csv')  
filepath.parent.mkdir(parents=True, exist_ok=True) 
dfProductosBS.to_csv(filepath)
```
